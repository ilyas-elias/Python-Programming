# Numpy X Pillow

```python
from PIL import Image
import numpy as np

img = np.array(Image.open('cat.jpg')) 
print(img.dtype, img.shape)  # Prints "uint8 (400, 248, 3)"

display(Image.fromarray(np.uint8(img)))

# We can tint the image by scaling each of the color channels
# by a different scalar constant. The image has shape (400, 248, 3);
# we multiply it by the array [1, 0.95, 0.9] of shape (3,);
# numpy broadcasting means that this leaves the red channel unchanged,
# and multiplies the green and blue channels by 0.95 and 0.9
# respectively.
img_tinted = img * [1, 0.95, 0.9]


# Resize the tinted image to be 300 by 300 pixels.
img_tinted = Image.fromarray(np.uint8(img_tinted)).resize((300, 300))

display(Image.fromarray(np.uint8(img_tinted)))
```

```python
import numpy as np
from PIL import Image

im_gray = np.array(Image.open('cat.jpg').convert('L'))      # Convert to grayscale (L - Luminance)
print(type(im_gray))

#display(Image.fromarray(np.uint8(im_gray)))

thresh = 128
im_bool = im_gray > thresh
print(im_bool)

maxval = 255

im_bin = (im_gray > thresh) * maxval
print(im_bin)

display(Image.fromarray(np.uint8(im_bin)))
```
